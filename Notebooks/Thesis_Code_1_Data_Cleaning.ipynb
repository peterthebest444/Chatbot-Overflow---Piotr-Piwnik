# 3. Data Engineering 

## 3.1 Add column with week number

```python
merged_df['Week_of_Year'] = merged_df['Post_Creation_Date'].dt.isocalendar().week
```

## 3.2 Separate the tags

#### Exploration

```python
def extract_tags(row, tag_index):
    tag_string = row['Post_Tags']
    tags = tag_string.strip().split('><')
    tags[0] = tags[0][1:]
    tags[-1] = tags[-1][:-1]

    if tag_index < len(tags):
        return tags[tag_index]
    else:
        return None

# Apply the extract_tags function for each tag index
merged_df['tag_1'] = merged_df.apply(lambda row: extract_tags(row, 0), axis=1)
merged_df['tag_2'] = merged_df.apply(lambda row: extract_tags(row, 1), axis=1)
merged_df['tag_3'] = merged_df.apply(lambda row: extract_tags(row, 2), axis=1)

merged_df = merged_df.drop(columns=['Post_Tags'])
```

## 3.3 Add the counts and group the data on forum-day basis

```python
merged_df.info()
```

```python
grouped_df = merged_df.groupby(['Forum', merged_df['Post_Creation_Date'].dt.date]).agg(
    Post_View_Count=('Post_View_Count', 'sum'),
    Post_Score=('Post_Score', 'sum'),
    Post_Answer_Count=('Post_Answer_Count', 'sum'),
    Response_Scores=('Response_Scores', 'sum'),
    Post_Author=('Post_Author', 'nunique'),
    Answer_Author=('Answer_Author', 'nunique'),
    Post_Count=('Post_Title', 'count'),
    Week_of_Year=('Week_of_Year', 'nunique'),
    tag_1_most_common=('tag_1', lambda x: x.mode().iloc[0] if not x.mode().empty else None),
    tag_2_most_common=('tag_2', lambda x: x.mode().iloc[0] if not x.mode().empty else None),
    tag_3_most_common=('tag_3', lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    
).reset_index()

```
